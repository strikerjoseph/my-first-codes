```python
# convertir les jours en semaines, mois et années
jours=int(input('nombre de jours:'))
sem= int(jours/7) 
j1=(jours%7)
mois= int(jours/30)
j2=jours%30
année= int(jours/365) 
jour= int(jours%365)
print (jours, 'jours equivaux à:\n\t',sem,'semaines et',j1,'jours\n\t\t', mois, 'mois et',j2,'jours\n\t\t\t',année,'années plus',jour,'jours' )
```

```python
# convertir en métre

mètre=float(input('distance en mètre:'))
kilo= mètre * 0.001
hecto= mètre * 0.01
deca= mètre * 0.1
metre=mètre
deci= mètre * 10
centi=mètre * 100
milli=mètre * 1000
print(mètre, 'm égale à', kilo,'Km')
print(mètre, 'm égale à', hecto,'Hm')
print(mètre, 'm égale à', deca,'Decam')
print(mètre, 'm égale à', metre,'m')
print(mètre, 'm égale à', deci,'dm')
print(mètre, 'm égale à', centi,'cm')
print(mètre, 'm égale à', milli,'mm')
```

```python
# Determiner le nombre de pizza a acheter :
# quatre personnes veulent commander des pizza, chaque personne souhaite manger 4 parts.
# la boutique locale de pizza fait des pizza de 6 parts
# quel est le nombre minimun de pizza à commander?

nombre_de_part=4*4
nombre_de_pizza=int((nombre_de_part+4)/6)
part_restant=(nombre_de_pizza*6)-nombre_de_part
print('il faudra commander ',nombre_de_pizza, 'Pizza, et il restera ', part_restant, 'parts')
```

```python
#les valeurs Booleenes

x= input('valeur de x=')
y=input('valeur de y=')
print('x=',x,'et','y=',y)
print('est-ce-que x est superieur à y ?:', x>y)
print('est-ce-que x est inferieur à y ? :', x<y)
print('est-ce-que x est égale à y ?:',x==y)
print('est-ce-que x est different de y ?:',x!=y)
print ('est-ce-que x est superieur ou égale à y ?:',x>=y)
print('est-ce-que x est inferieur ou égale à y ?:',x<=y,'\n\n')
```

```python
###### combinaison de valeurs Booleenne #####
a=int(input('a='))
c=int(input('c='))

print('a et c inferieur à 100:', a<100 and c<100)
print('a ou c inferieur à 22:',a<22 or c<22)
print('a et c inferieur à 22:', a<22 and c<22,end='\n\n')

#print(not True) la negation de notre affirmation

print('not(a et c inferieur à 100):', not(a<100 and c<100))
print('not(a ou c inferieur à 22):',not(a<22 or c<22))
print('not(a et c inferieur à 22):', not(a<22 and c<22),end='\n\n')
```

```python
# question 1:
# write a code that ask the user to input a number between 1 and 5 inclusive.
# The code will take the integer value and print out the string value.

c = int(input('entrer un numero de 1 à 5: '))
if c==1:
    print('Un')
elif c==2:
    print('Deux')
elif c==3:
    print('Trois')
elif c==4:
    print('Quatre')
elif c==5:
    print('Cinq')
else:
    print ('Out of Range')
```
