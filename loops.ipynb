## "for" loop ##

```python
for i in range(21):
    print(i,end=',')
```

```python
for T in range(0,124,4):
    print(T,end=',') 
```

```python
for T in range(124,0,-4):
    print(T,end=',') 
```

```python
word= 'blackcat'
for i in word:
    print(i,end=',')
```

**** #### "while" loop ##### ****

```python
n=10
while n>0:
    print(n, end=' ')
    n=n-1
```

```python
age=int(input('age des eleves de la classe,(ECRIRE UN NOMBRE NEGATIF POUR STOPPER) = '))
ages=[]
while age>0:
    ages.append(age)
    age=int(input('next age:>>  '))
print('l\'age des eleves de la classe:',ages)
```

```python
conte=0
liste=[]
nom=input('SVP entrez le nom: ')
while nom!='JEN Z':
    conte+=1
    liste.append(nom)
    print(f'{nom} a été ajouté.\n')
    nom=input('nom suivant:>')
print(f'il y a {conte} etudiants dans la classe, il s\'agit de {liste}')
```
