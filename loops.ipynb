```python
age=int(input('age des eleves de la classe,(ECRIRE UN NOMBRE NEGATIF POUR STOPPER) = '))
ages=[]
while age>0:
    ages.append(age)
    age=int(input('next age:>>  '))
print('l\'age des eleves de la classe:',ages)
```

```python
conte=0
liste=[]
nom=input('SVP entrez le nom: ')
while nom!='JEN Z':
    conte+=1
    liste.append(nom)
    print(f'{nom} a été ajouté.\n')
    nom=input('nom suivant:>')
print(f'il y a {conte} etudiants dans la classe, il s\'agit de {liste}')
```

```python
n=150

for i in range(1,n+1):
    if i%3==0 and i%5==0:
        print (i,'FizzBuzz')
    elif i%3==0:
        print(i,'Fizz')
    elif i%5==0:
        print(i,'Buzz')
    else:
        print(i)
```

# enter two numbers between 1 and 150. then count 
# from the lower number to the higher number.print the result to the screen

```python
a= int(input('Please enter a number between 1-150:>'))
b= int(input('Please enter a number between 1-150:>'))

while a<0 or b<0  and a>150 or b>150 :
    print ('Number must be different values between 1-150, please try again')
    a= int(input('Please enter a number between 1-150:>'))
    b= int(input('Please enter a number between 1-150:>'))
if a<b:
    for i in range(a,b+1):
        print(i, end=' ')
else:
    for i in range(b,a+1):
        print(i,end=' ')
```
